```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os


df = pd.read_csv("etf_prices.csv", parse_dates=["Date"], index_col="Date")

# Obliczenie dziennych zwrot√≥w
returns = df.pct_change().dropna()

# Podstawowe statystyki
stats = returns.describe()

```

```python
# Macierz korelacji
correlation_matrix = returns.corr()

# Wizualizacja zwrot√≥w
plt.figure(figsize=(12, 6))
returns.plot(figsize=(12, 6), title="Dzienny zwrot ETF-√≥w")
plt.ylabel("Zwrot dzienny")
plt.legend(loc="upper right")
plt.grid()
plt.show()

# Wizualizacja korelacji
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Macierz korelacji zwrot√≥w ETF-√≥w")
plt.show()

```
